```python
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
```

```python
df=pd.read_csv('NY-House-Dataset.csv')
df
```

```python
df.nunique()
```

```python
df=df.drop(columns=['BROKERTITLE','TYPE','ADDRESS','STATE','LOCALITY','MAIN_ADDRESS','ADMINISTRATIVE_AREA_LEVEL_2','SUBLOCALITY','STREET_NAME','LONG_NAME','FORMATTED_ADDRESS','LATITUDE','LONGITUDE'])
df
```

```python
X = df.drop(columns=(['PRICE']))
Y = df['PRICE']
X 
Y
```

```python
df = pd.get_dummies(df, columns=['BATH', 'BEDS'], dtype="int")
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test,Y_train, Y_test = train_test_split(X, Y, test_size=0.8)
```

```python
from sklearn.linear_model import LinearRegression
model1= LinearRegression()
model1.fit(X_train, Y_train)
```

```python
model1.score(X_train,Y_train)
```

```python
model1.score(X_test,Y_test)
```

```python
print("The Accuracy is  ",int ((model1.score(X_train,Y_train))*100),'%')
```

```python
from sklearn.linear_model import Lasso
```

```python
reg2 = Lasso(alpha=75,max_iter=100,tol=0.4)
```

```python
reg2.fit(X_train,Y_train)
```

```python
reg2.score(X_train,Y_train)
```

```python
reg2.score(X_test,Y_test)
```

```python
reg2 = Lasso(alpha=25,max_iter=1000,tol=0.4)
```

```python
reg2.fit(X_train,Y_train)
```

```python
reg2.score(X_train,Y_train)
```

```python
reg2.score(X_test,Y_test)
```
